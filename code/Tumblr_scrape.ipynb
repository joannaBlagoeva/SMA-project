Creating a folder if one does not exist

```python
folder_name = Path(date_of_tags)
folder_name.mkdir(parents=True, exist_ok=True)
```

```python
def scrape_tag(tag, file_name, date_range):
    print(f"Writing in file {file_name}")
    with open(file_name, 'w', encoding='utf-8') as output_file:

    #write header of file
        output_file.write("note_count,primary_tag,secondary_tags,blog_name,post_body,date\n")
        for date_time in date_range.to_list():
   
            print(date_time)
            dt = time.strptime(str(date_time), "%Y-%m-%d %H:%M:%S")

        #setup epoch time
            epoch_time = int(time.mktime(dt))

        #make api call for tag before date -> result 20 posts
            date_posts = client.tagged(str(tag), before=epoch_time, filter='text')

        #fetch important data per post
            for dp in date_posts:
                
                post_body=""
                is_text=False
                if 'body' in dp:
                    post_body=html.escape(str(dp['body']))
                    if post_body != "":
                        is_text=True

                
                sec_tags=list(map(lambda x: x.lower(), dp['tags']))
                if tag.lower() in sec_tags:
                    sec_tags.remove(tag.lower())
                sec_tags=str(sec_tags).replace(','," ")

            
                if is_text:
                    text_body=post_body.replace(",","").replace("[","").replace("]","").replace("'","").replace("\n"," ").replace("\t"," ").replace("\""," ")
                    output_file.write(f"{dp['note_count']},{tag},{sec_tags},{dp['blog_name']},{text_body},{dp['date']}\n")
```

```python
for tag in read_tags:
    print(f"sraping for tag {tag}")
    file_name = folder_name / f"{tag.replace(" ","_")}.csv"
    scrape_tag(tag, file_name, search_date_range)
```
